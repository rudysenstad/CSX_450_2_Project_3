```r
library(dplyr, warn.conflicts = FALSE)
library(moments)
library(ggplot2)
library(repr)
source('multiplot.r')
options(repr.plot.width=10, repr.plot.height=2) 
```

```r
housing_data <- data.frame(read.table(file='../data/housing.data', header=FALSE, sep=""))
colnames(housing_data) <- c('CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV')
housing_data$MEDV <- NULL
housing_data$CHAS = factor(housing_data$CHAS)
housing_features = Filter(is.numeric, housing_data)
```

# Data Conditioning

### Prepare distribution plots that show the mean and median for each feature

```r
original <- function (df, feature1, feature2, label) {
    original_feature1 <- as.vector(df[[feature1]])
    p1 <- qplot(original_feature1, geom="histogram", bins=200, alpha=I(.4),)+
        geom_vline(aes(xintercept=mean(original_feature1)), color="red", linetype="dashed", size=1)+
        geom_vline(aes(xintercept=median(original_feature1)), color="blue", linetype="dashed", size=1)+
        ggtitle(paste(label, " Distribution of", feature1))
    original_feature2 <- as.vector(df[[feature2]])
    p2 <- qplot(original_feature2, geom="histogram", bins=200, alpha=I(.4),)+
        geom_vline(aes(xintercept=mean(original_feature2)), color="red", linetype="dashed", size=1)+
        geom_vline(aes(xintercept=median(original_feature2)), color="blue", linetype="dashed", size=1)+
        ggtitle(paste(label, " Distribution of", feature2))
    multiplot(p1, p2, cols = 2)
}
```

```r
original(housing_features, 'CRIM','ZN', 'Original')
original(housing_features, 'INDUS','NOX', 'Original')
original(housing_features, 'RM','AGE', 'Original')
original(housing_features, 'DIS','RAD', 'Original')
original(housing_features, 'TAX','PTRATIO', 'Original')
original(housing_features, 'B','LSTAT', 'Original')
```

### Discuss how we can assess skew from these plots

1. Positive skew - the median is to the right (positive direction on the X axis) of the mean
2. Negative skew - the median is to the left (negative direction on the X axis) of the mean

With the red line being mean and the blue line being median we can comment on each feature's skewness
    1. CRIM - Negative skew
    2. ZN - Negative skew
    3. INDUS - Negative skew
    4. NOX - Negative skew
    5. RM - Very slight negative skew, the mean and median are very close
    6. AGE - Positive skew
    7. DIS - Negative skew
    8. RAD - Negative skew
    9. TAX - Negative skew
    10. PTRATIO - Positive skew
    11. B - Positive skew
    12. LSTAT - Negative skew


### Standardize the data

```r
housing_features_scaled = data.frame(scale(housing_features))
stats_sc = data.frame(feature=colnames(housing_features_scaled))
stats_sc['mean_'] = sapply(housing_features_scaled, mean)
stats_sc['sd_'] = sapply(housing_features_scaled, sd)
stats_sc['skewness_'] = sapply(housing_features_scaled, skewness)
stats_sc['kurtosis_'] = sapply(housing_features_scaled, kurtosis)
stats_sc
```

### Prepare distribution plots of the standardized data and discuss how the plots show the transformation that's been applied

```r
original(housing_features_scaled, 'CRIM','ZN','Scaled')
original(housing_features_scaled, 'INDUS','NOX','Scaled')
original(housing_features_scaled, 'RM','AGE','Scaled')
original(housing_features_scaled, 'DIS','RAD','Scaled')
original(housing_features_scaled, 'TAX','PTRATIO','Scaled')
original(housing_features_scaled, 'B','LSTAT','Scaled')
```

The value of the features (on the X-axis) have been modified to scales that are easier to refer to and compare amongst each other. Most of the values of all features are between -4 and 4.

### Apply a log transformation to the original data, then standardize or standard scale it

```r
skewness_ = data.frame(feature=colnames(housing_features))
skewness_['skew_'] = sapply(housing_features, skewness)

housing_features_log_df = log(housing_features)

skewness_['skew_log_'] = sapply(housing_features_log_df, skewness)
skewness_
```

```r
housing_features_log_scaled_df = data.frame(scale(housing_features_log_df))
skewness_['skew_log_sc_'] = sapply(housing_features_log_scaled_df, skewness)
skewness_
```

### Prepare distribution plots of the transformed data and discuss how the plots show the transformation that's been applied

```r
original(housing_features_log_scaled_df, 'CRIM','ZN','Log & Scaled')
original(housing_features_log_scaled_df, 'INDUS','NOX','Log & Scaled')
original(housing_features_log_scaled_df, 'RM','AGE','Log & Scaled')
original(housing_features_log_scaled_df, 'DIS','RAD','Log & Scaled')
original(housing_features_log_scaled_df, 'TAX','PTRATIO','Log & Scaled')
original(housing_features_log_scaled_df, 'B','LSTAT','Log & Scaled')
```

Taking the log of the data can have the effect of pulling in the outliers closer to the mean. This in turn will reduce the 3rd moment of probability, also known as the skewness. In this dataset, 7 of the features have reduced skewness after taking the log. This is depicted in the plots by the closeness of the mean and median when compared to the distribution plots of the original data. Also, the area of bar plots on each side of the median starts to look more evenly distributed.
