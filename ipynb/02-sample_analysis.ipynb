```r
library(moments)
library(dplyr, warn.conflicts = FALSE)
library(ggplot2)
library(repr)
library(reshape2)
```

```r
housing_data <- data.frame(read.table(file='../data/housing.data', header=FALSE, sep=""))
# colnames(housing_data) <- c('crime','zoned','industrial','river','nox','rooms','forties',
#                             'work_distance','radial_highway','tax_rate',
#                             'student_teacher','african-american','lower_status','median_value')
colnames(housing_data) <- c('CRIM','ZN','INDUS','CHAS','NOX','RM','AGE','DIS','RAD','TAX','PTRATIO','B','LSTAT','MEDV')
housing_data$MEDV <- NULL
housing_data$CHAS = factor(housing_data$CHAS)
housing_features = Filter(is.numeric, housing_data)
```

### Select a small sample (3-5 points) of the dataset

```r
sample_df = data.frame(sample_n(housing_features, 5))
sample_df
```

### Compare this sample to the dataset as a whole

```r
stats = data.frame(feature=colnames(housing_features))
stats['mean_'] = sapply(housing_features, mean)
stats['sd_'] = sapply(housing_features, sd)
stats['skewness_'] = sapply(housing_features, skewness)
stats
```

```r
sample_stats_df = data.frame(feature=colnames(sample_df))
sample_stats_df['mean_'] = sapply(sample_df, mean)
sample_stats_df['sd_'] = sapply(sample_df, sd)
sample_stats_df['skewness_'] = sapply(sample_df, skewness)
sample_stats_df
```

### Convert the samples into Z-scores

```r
z_scores_ = (t(sample_df) - stats$mean_)/stats$sd_
z_df = data.frame(z_scores_)
z_df['feature'] <- colnames(sample_df)
z_df
```

### Prepare a heatmap and a barplot of your samples and of the sample z-scores

```r
options(repr.plot.width=10, repr.plot.height=6)
p = ggplot(data=melt(z_df), aes(x=feature, y=value, fill = variable)) + geom_bar(stat = "identity", width=0.75, position = "dodge")
p
# p + coord_flip()
```

```r
options(repr.plot.width=10, repr.plot.height=6)
thisone = data.frame(t(sample_df))
thisone['feature'] = colnames(sample_df)
p = ggplot(data=melt(thisone), aes(x=feature, y=value, fill = variable)) + geom_bar(stat = "identity", width=0.75, position = "dodge")
p + coord_flip()
```

```r
options(repr.plot.width=6, repr.plot.height=6)
pairs(t(z_scores_)); pairs(sample_df)
```
